```python
import pandas as pd
```

```python
df = pd.read_csv(
    "../data/raw/stores_sales_forecasting.csv",
    encoding="latin1"
)

df.head()
```

## Standardize Column Names

```python
df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(" ", "_")
)

df.columns
```

```python
df.columns = df.columns.str.replace("-", "_")
df.columns
```

```python
df["order_date"] = pd.to_datetime(df["order_date"])
df["ship_date"] = pd.to_datetime(df["ship_date"])
```

```python
df[["order_date", "ship_date"]].head()
```

```python
columns_to_drop = [
    "row_id",
    "order_id",
    "customer_id",
    "customer_name",
    "postal_code",
    "product_id",
    "product_name"
]

df = df.drop(columns=columns_to_drop)
df.head()
```

```python
df = df.sort_values("order_date")
df.reset_index(drop=True, inplace=True)
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df.to_csv(
    "../data/cleaned/cleaned_sales_data.csv",
    index=False
)
```
