```python
import pandas as pd
```

```python
monthly_sales = pd.read_csv(
    "../data/processed/monthly_sales_trends.csv",
    parse_dates=["order_date"]
)

category_monthly_sales = pd.read_csv(
    "../data/processed/category_monthly_sales.csv",
    parse_dates=["order_date"]
)

forecast = pd.read_csv(
    "../data/processed/next_month_forecast.csv",
    parse_dates=["month"]
)

monthly_sales.head(), category_monthly_sales.head(), forecast.head()
```

```python
latest_month = monthly_sales.iloc[-1]

kpi_df = pd.DataFrame({
    "kpi": [
        "Latest Month Sales",
        "Month-over-Month Growth (%)",
        "3-Month Rolling Average",
        "Next Month Forecast"
    ],
    "value": [
        round(latest_month["sales"], 2),
        round(latest_month["mom_growth_pct"], 2),
        round(latest_month["rolling_3_month_avg"], 2),
        round(forecast["forecast_sales"].iloc[0], 2)
    ]
})

kpi_df
```

```python
monthly_sales.to_csv(
    "../powerbi/monthly_sales_trends_pbi.csv",
    index=False
)

category_monthly_sales.to_csv(
    "../powerbi/category_monthly_sales_pbi.csv",
    index=False
)

kpi_df.to_csv(
    "../powerbi/kpi_metrics_pbi.csv",
    index=False
)

forecast.to_csv(
    "../powerbi/forecast_pbi.csv",
    index=False
)
```
