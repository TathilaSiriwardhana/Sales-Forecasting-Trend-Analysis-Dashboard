## Import Libraries & Load Data

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(
    "../data/raw/stores_sales_forecasting.csv",
    encoding="latin1"
)

df.head()
```

## Key Columns for Analysis

Order Date: Time series reference

Sales: Revenue metric

Category / Sub-Category: Product-level trends

Region / Segment: Optional slicing

Quantity, Discount, Profit: Supporting metrics

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Ship Date"] = pd.to_datetime(df["Ship Date"])
```

```python
df[["Order Date", "Ship Date"]].head()
```

```python
df["Order Date"].min(), df["Order Date"].max()
```

```python
df["Sales"].describe()
```

```python
df["Sales"].hist(bins=30)
```

```python
df.groupby("Category")["Sales"].sum().sort_values(ascending=False)
```

```python
monthly_sales_preview = (
    df.set_index("Order Date")
      .resample("M")["Sales"]
      .sum()
)

monthly_sales_preview.head()
```

## Initial Findings

Dataset is clean with no missing values

Time span is sufficient for trend analysis

Sales data supports monthly resampling

Categories provide meaningful segmentation
